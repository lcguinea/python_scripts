```python
# Busca un track de Spotify y actualiza los metadatos de un archivo MP3

import requests
from requests.auth import HTTPBasicAuth
import json
from mutagen.easyid3 import EasyID3
from mutagen.mp3 import MP3
from mutagen.id3 import ID3, APIC

# Configura tus credenciales de Spotify
client_id = 'YOUR_CLIENT_ID'
client_secret = 'YOUR_CLIENT_SECRET'

# URL del track de Spotify
spotify_url = input('Ingresa el URL del track de Spotify: ')

# Ruta al archivo MP3
mp3_file_path = input('Ingresa la ruta al archivo MP3: ')

# Obtener el token de acceso
auth_response = requests.post(
    'https://accounts.spotify.com/api/token',
    data={'grant_type': 'client_credentials'},
    auth=HTTPBasicAuth(client_id, client_secret)
)
auth_response_data = auth_response.json()
access_token = auth_response_data['access_token']

# Obtener el ID del track desde el URL
track_id = spotify_url.split('/')[-1].split('?')[0]

# Obtener los datos del track
headers = {
    'Authorization': f'Bearer {access_token}'
}
track_response = requests.get(f'https://api.spotify.com/v1/tracks/{track_id}', headers=headers)
track_data = track_response.json()

# Extraer los datos necesarios
artist_names = ', '.join([artist['name'] for artist in track_data['artists']])
album_name = track_data['album']['name']
track_name = track_data['name']
release_date = track_data['album']['release_date']
disc_number = track_data['disc_number']
track_number = track_data['track_number']
isrc = track_data['external_ids']['isrc']
cover_url = track_data['album']['images'][0]['url']

# Imprimir los datos obtenidos
print(f'Artist: {artist_names}')
print(f'Album: {album_name}')
print(f'Track name: {track_name}')
print(f'Release date: {release_date}')
print(f'Disc number: {disc_number}')
print(f'Track number: {track_number}')
print(f'ISRC: {isrc}')
print(f'Cover URL: {cover_url}')

# Cargar el archivo MP3
audio = MP3(mp3_file_path, ID3=EasyID3)

# Actualizar los metadatos
audio['artist'] = artist_names
audio['album'] = album_name
audio['title'] = track_name
audio['date'] = release_date
audio['discnumber'] = str(disc_number)
audio['tracknumber'] = str(track_number)
audio['isrc'] = isrc

# Guardar los cambios
audio.save()

# AÃ±adir la imagen de portada
audio = MP3(mp3_file_path, ID3=ID3)
response = requests.get(cover_url)
audio.tags.add(
    APIC(
        encoding=3,  # 3 is for utf-8
        mime='image/jpeg',  # image mime type
        type=3,  # 3 is for the cover image
        desc=u'Cover',
        data=response.content
    )
)

# Guardar los cambios
audio.save()

print('Metadatos actualizados correctamente.')
```
