```python
# Find all unique file formats in a directory and its subdirectories

import os

# Define the root folder to scan
root_folder = '/Users/luisguinea/Library/CloudStorage/GoogleDrive-luis.guinea@recmusica.com/Mi unidad/Peermusic/Nacho Campillo'

# Set to store unique file formats
file_formats = set()

# Traverse all directories and subdirectories
for dirpath, dirnames, filenames in os.walk(root_folder):
    for filename in filenames:
        # Get the file extension
        file_extension = os.path.splitext(filename)[1]
        if file_extension:
            file_formats.add(file_extension.lower())

# Print the list of unique file formats
print("File formats found:")
for file_format in sorted(file_formats):
    print(file_format)
```

```python
# Convert all audio files to mp3 format and move mp3 files to a destination folder

import os
from pydub import AudioSegment

# Define the root folder to scan
root_folder = '/Users/luisguinea/Library/CloudStorage/GoogleDrive-luis.guinea@recmusica.com/Mi unidad/Peermusic/Nacho Campillo'

# Define the destination folder for mp3 files
destination_folder = '/Users/luisguinea/Downloads/mp3'

# Ensure the destination folder exists
os.makedirs(destination_folder, exist_ok=True)

# Traverse all directories and subdirectories
for dirpath, dirnames, filenames in os.walk(root_folder):
    for filename in filenames:
        file_path = os.path.join(dirpath, filename)
        file_extension = os.path.splitext(filename)[1].lower()

        if file_extension in ['.mp4', '.wav']:
            # Define the output mp3 file path
            mp3_filename = os.path.splitext(filename)[0] + '.mp3'
            mp3_path = os.path.join(destination_folder, mp3_filename)

            # Convert the file to mp3
            audio = AudioSegment.from_file(file_path, format=file_extension[1:])
            audio.export(mp3_path, format='mp3')
            print(f'Converted: {file_path} to {mp3_path}')

        elif file_extension == '.mp3':
            # Move the mp3 file to the destination folder
            destination_path = os.path.join(destination_folder, filename)
            os.rename(file_path, destination_path)
            print(f'Moved: {file_path} to {destination_path}')

print("Conversion and moving completed.")
```
