```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table
from PIL import Image

# File path
input_file = '/Users/luisguinea/Library/CloudStorage/GoogleDrive-luis.guinea@recmusica.com/Mi unidad/Peermusic/Cata패logo Pep\'s Music/sorted_track_metadata.csv'
output_image = '/Users/luisguinea/Library/CloudStorage/GoogleDrive-luis.guinea@recmusica.com/Mi unidad/Peermusic/Cata패logo Pep\'s Music/top_20_tracks.jpg'

# Load the sorted CSV file into a pandas dataframe
df = pd.read_csv(input_file, sep=',')

# Ensure there are no leading/trailing spaces in column names
df.columns = df.columns.str.strip()

# Select the top 20 tracks
top_20_df = df.head(20)

# Remove the 'Spotify Link' column
top_20_df = top_20_df.drop(columns=['Spotify Link'])

# Plotting the table
fig, ax = plt.subplots(figsize=(14, 10))  # Increase the size of the figure
ax.xaxis.set_visible(False)  # Hide the x-axis
ax.yaxis.set_visible(False)  # Hide the y-axis
ax.set_frame_on(False)  # Hide the frame
plt.subplots_adjust(left=0.1, top=0.9, right=0.9, bottom=0.1)  # Adjust the plot's margins

# Create a table and add it to the plot
tbl = table(ax, top_20_df, loc='center', cellLoc='center', colWidths=[0.2]*len(top_20_df.columns))

# Style the headers
tbl.auto_set_font_size(False)
tbl.set_fontsize(8)  # Adjust the font size for better readability
tbl.scale(1.2, 1.5)  # Scale the table for more spacing

# Make headers bold
for key, cell in tbl.get_celld().items():
    if key[0] == 0:  # Header cells
        cell.set_text_props(fontweight='bold')

# Save the plot as a JPG file
plt.savefig(output_image, format='jpg', bbox_inches='tight', dpi=300)  # Increase dpi for better quality

# Optionally, convert the image to ensure it's in JPG format
img = Image.open(output_image)
img.convert('RGB').save(output_image)

print(f'Top 20 tracks image saved to {output_image}')
```

```python
import pandas as pd

# File path
input_file = '/Users/luisguinea/Library/CloudStorage/GoogleDrive-luis.guinea@recmusica.com/Mi unidad/Peermusic/Cata패logo Pep\'s Music/sorted_track_metadata.csv'
output_excel = '/Users/luisguinea/Library/CloudStorage/GoogleDrive-luis.guinea@recmusica.com/Mi unidad/Peermusic/Cata패logo Pep\'s Music/top_20_tracks.xlsx'

# Load the sorted CSV file into a pandas dataframe
df = pd.read_csv(input_file, sep=',')

# Ensure there are no leading/trailing spaces in column names
df.columns = df.columns.str.strip()

# Select the top 20 tracks
top_20_df = df.head(20)

# Remove the 'Spotify Link' column
top_20_df = top_20_df.drop(columns=['Spotify Link'])

# Write the data to an Excel file
with pd.ExcelWriter(output_excel, engine='openpyxl') as writer:
    top_20_df.to_excel(writer, sheet_name='Top 20 Tracks', index=False)

    # Access the workbook and sheet to format headers
    workbook = writer.book
    worksheet = writer.sheets['Top 20 Tracks']

    # Apply bold formatting to headers
    for cell in worksheet[1]:
        cell.font = cell.font.copy(bold=True)

print(f'Top 20 tracks Excel report saved to {output_excel}')
```

```python
import pandas as pd

# File paths
input_file = '/Users/luisguinea/Library/CloudStorage/GoogleDrive-luis.guinea@recmusica.com/Mi unidad/Peermusic/Cata패logo Pep\'s Music/sorted_track_metadata.csv'
output_excel_all = '/Users/luisguinea/Library/CloudStorage/GoogleDrive-luis.guinea@recmusica.com/Mi unidad/Peermusic/Cata패logo Pep\'s Music/all_tracks_with_spotify_link.xlsx'

# Load the CSV file into a pandas dataframe
df = pd.read_csv(input_file, sep=',')

# Ensure there are no leading/trailing spaces in column names
df.columns = df.columns.str.strip()

# Reorder columns to move 'Spotify Link' to the end
columns = [col for col in df.columns if col != 'Spotify Link'] + ['Spotify Link']
df = df[columns]

# Write the data to an Excel file
with pd.ExcelWriter(output_excel_all, engine='openpyxl') as writer:
    df.to_excel(writer, sheet_name='All Tracks', index=False)

    # Access the workbook and sheet to format headers
    workbook = writer.book
    worksheet = writer.sheets['All Tracks']

    # Apply bold formatting to headers
    for cell in worksheet[1]:
        cell.font = cell.font.copy(bold=True)

print(f'All tracks Excel report saved to {output_excel_all}')
```
