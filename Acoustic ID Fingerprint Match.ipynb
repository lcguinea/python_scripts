```python
import os
import acoustid
import tinytag

API_KEY = 'YOUR_API_KEY'

def fingerprint_and_get_metadata(file_path):
    # Generate fingerprint using pyacoustid
    duration, fingerprint = acoustid.fingerprint_file(file_path)
    result = acoustid.lookup(API_KEY, fingerprint, duration)
    
    # Get metadata using tinytag
    tag = tinytag.TinyTag.get(file_path)
    
    return result, tag

def scan_folder(folder_path):
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.mp3'):
                file_path = os.path.join(root, file)
                result, tag = fingerprint_and_get_metadata(file_path)
                print(f"File: {file_path}")
                print(f"Metadata: {tag}")
                print(f"Fingerprint Result: {result}")

# Replace 'your_folder_path' with the path to your folder of MP3 files
scan_folder('/Users/luisguinea/Downloads/MAGIK 2.0 Mp3')
```

```python
import acoustid
import chromaprint

# Your AcousticID API key (replace with your actual API key)
API_KEY = "WwRhXhcqQ5"

def analyze_mp3(file_path):
    try:
        # Generate a fingerprint and duration using Chromaprint
        duration, fingerprint = acoustid.fingerprint_file(file_path)
        print(f"Fingerprint generated. Duration: {duration} seconds")

        # Identify the track using AcousticID
        results = acoustid.lookup(API_KEY, fingerprint, duration)
        
        # Debugging: Print the full response to inspect its structure
        print("Full API Response:")
        print(results)

        # Print the results
        if 'results' in results:
            for result in results['results']:
                score = result.get('score', 'N/A')
                recording = result.get('recordings', [{}])[0]  # Get the first recording if available
                title = recording.get('title', 'Unknown Title')
                artist = recording.get('artists', [{'name': 'Unknown Artist'}])[0].get('name', 'Unknown Artist')
                recording_id = recording.get('id', 'Unknown ID')
                
                print(f"Title: {title}")
                print(f"Artist: {artist}")
                print(f"Recording ID: {recording_id}")
                print(f"Score: {score}")
                print()
        else:
            print("No results found.")

    except acoustid.AcoustidError as e:
        print(f"Error identifying track: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Usage example
if __name__ == "__main__":
    # Replace with the path to your MP3 file
    mp3_file_path = input("Enter the path to an MP3 file: ")
    analyze_mp3(mp3_file_path)
```
