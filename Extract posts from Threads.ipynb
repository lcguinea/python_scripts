```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import time
from docx import Document

# Path to the msedgedriver
edge_driver_path = '/Users/luisguinea/Downloads/edgedriver_mac64_m1/msedgedriver'

# Set up Edge options
options = Options()
# options.add_argument('--headless')  # Uncomment if you want to run in headless mode

# Initialize the Edge service and driver
service = Service(executable_path=edge_driver_path)
driver = webdriver.Edge(service=service, options=options)

# List to store extracted text
all_extracted_text = []

# Read URLs from the text file
with open('/Users/luisguinea/Desktop/Threads_links.txt', 'r') as file:
    urls = file.readlines()

try:
    for url in urls:
        url = url.strip()  # Remove any leading/trailing whitespace
        if not url:
            continue  # Skip empty lines

        # Open the specific thread post URL
        driver.get(url)
        
        # Wait for the page to load and the specific elements to be present
        WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.TAG_NAME, 'body'))
        )
        
        # Scroll down to load all posts if needed
        body = driver.find_element(By.TAG_NAME, 'body')
        body.send_keys(Keys.PAGE_DOWN)
        time.sleep(2)  # Wait for new content to load
        
        # Extract all <h1> and <span> elements with the specified class
        elements = driver.find_elements(By.CSS_SELECTOR, 'h1.x1lliihq.x1plvlek.xryxfnj.x1n2onr6.x193iq5w.xeuugli.x1fj9vlw.x13faqbe.x1vvkbs.x1s928wv.xhkezso.x1gmr53x.x1cpjm7i.x1fgarty.x1943h6x.x1i0vuye.xjohtrz.xo1l8bm.xp07o12.x1yc453h.xat24cr.xdj266r, span.x1lliihq.x1plvlek.xryxfnj.x1n2onr6.x193iq5w.xeuugli.x1fj9vlw.x13faqbe.x1vvkbs.x1s928wv.xhkezso.x1gmr53x.x1cpjm7i.x1fgarty.x1943h6x.x1i0vuye.xjohtrz.xo1l8bm.xp07o12.x1yc453h.xat24cr.xdj266r')
        
        # Collect the text of each element
        for element in elements:
            all_extracted_text.append(element.text)
            all_extracted_text.append('-' * 80)  # Separator for readability

finally:
    # Close the driver
    driver.quit()

# Create a new Document
doc = Document()

# Add a title
doc.add_heading('Extracted Posts', level=1)

# Add the extracted text to the document
for text in all_extracted_text:
    doc.add_paragraph(text)

# Save the document
doc.save('Extracted_Posts.docx')
```

```python
# Remove duplicates

from docx import Document

# Path to the existing document
input_path = '/Users/luisguinea/Documents/Pruebas/Extracted_Posts.docx'
output_path = '/Users/luisguinea/Documents/Pruebas/Cleaned_Extracted_Posts.docx'

# Open the existing document
doc = Document(input_path)

# Read all paragraphs and store them in a list
paragraphs = [para.text for para in doc.paragraphs]

# Remove duplicate paragraphs while maintaining order
seen = set()
unique_paragraphs = []
for para in paragraphs:
    if para not in seen:
        seen.add(para)
        unique_paragraphs.append(para)

# Create a new Document
cleaned_doc = Document()

# Add the unique paragraphs to the new document
for para in unique_paragraphs:
    cleaned_doc.add_paragraph(para)

# Save the cleaned document
cleaned_doc.save(output_path)

print(f"Cleaned document saved to {output_path}")
```
